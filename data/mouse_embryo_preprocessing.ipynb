```python
tree = dendropy.Tree.get(
      path=f'{data_path}/mouse/GSE117542_embryo2_BestTreeGreedy_with_childern.nwk',
      schema='newick')
tree.suppress_unifurcations()  # node before each leaf has leaf's barcode and leaf is only child
for edge in tree.edges():
  edge.length = 1
tree.write(path=f'{data_path}/preprocessed/GSE117542_embryo2_pruned.nwk', schema='newick')

print('Height', tree.max_distance_from_root())
degrees, counts = np.unique(list(map(lambda x: x.num_child_nodes(), tree.internal_nodes())), return_counts=True)
print('avg deg.', np.sum(degrees * counts) / np.sum(counts))
print('max deg.', degrees[-1])
leaves = np.array([t.label for t in tree.taxon_namespace])
print(len(leaves))
print(degrees)
print(counts)
leaves
```

## Gene expression pre-processing

```python
tree = dendropy.Tree.get(
      path=f'{data_path}/preprocessed/GSE117542_embryo2_pruned.nwk',
      schema='newick')
leaves = np.array([t.label for t in tree.taxon_namespace])

# R sparse matrix
print('counts:')
expression_adata_a = mmread(f'{data_path}/mouse/GSM3302827_embryo2a_SeuratNorm.mtx')
print(expression_adata_a.shape)
expression_adata_b = mmread(f'{data_path}/mouse/GSM3302828_embryo2b_SeuratNorm.mtx')
print(expression_adata_b.shape)

print('\ngenes:')
gene_names_a = pd.read_table(f'{data_path}/mouse/GSM3302827_embryo2a_SeuratGenes.tsv', header=None)[0].values
print(len(gene_names_a))
print(gene_names_a)
gene_names_b = pd.read_table(f'{data_path}/mouse/GSM3302828_embryo2b_SeuratGenes.tsv', header=None)[0].values
print(len(gene_names_b))
print(gene_names_b)

print('\nbarcodes:')
barcodes_a = pd.read_table(f'{data_path}/mouse/GSM3302827_embryo2a_SeuratBarcodes.tsv', header=None)[0]
barcodes_a = np.array([x + '-1' for x in barcodes_a]) # to match the newick file
print(len(barcodes_a))
print(barcodes_a)
barcodes_b = pd.read_table(f'{data_path}/mouse/GSM3302828_embryo2b_SeuratBarcodes.tsv', header=None)[0]
barcodes_b = np.array([x + '-2' for x in barcodes_b]) # to match the newick file
print(len(barcodes_b))
print(barcodes_b)
```

```python
is_valid_a = np.isin(barcodes_a, leaves)
print(np.sum(is_valid_a))
is_valid_b = np.isin(barcodes_b, leaves)
print(np.sum(is_valid_b))

expression_df_a = pd.DataFrame(expression_adata_a.toarray()[:, is_valid_a].T,
                             index=barcodes_a[is_valid_a], columns=gene_names_a)
expression_df_b = pd.DataFrame(expression_adata_b.toarray()[:, is_valid_b].T,
                             index=barcodes_b[is_valid_b], columns=gene_names_b)
expression_df = pd.concat((expression_df_a, expression_df_b))
expression_df = expression_df.fillna(0)
expression_df
```

```python
# save cell names
expression_df.to_csv(f'{data_path}/preprocessed/GSE117542_embryo2_cells.txt',
                     columns=[], header=False, index=True)
```

```python
# remove rare genes
print(expression_df.shape)
expression_df = expression_df.iloc[:, ((expression_df > 0).sum(axis=0) >= 10).values.flatten()]
print(expression_df.shape)
# skip L1 and log normalization as original dataset is already normalized

keep_gene_names = expression_df.columns
gene_counts = expression_df.values
df = pd.DataFrame(gene_counts)
df.columns = keep_gene_names
df.to_csv(f'{data_path}/preprocessed/GSE117542_embryo2_normalized_log_counts.txt', index=False)
df
```

## Cell types

```python
# adapted from how cell types are extraced in https://github.com/deto/Hotspot_Analysis/blob/master/Notebooks/Lineage_Tutorial.ipynb

import gzip
from scipy.spatial.distance import cdist
import seaborn as sns
from umap import UMAP
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

with gzip.open(f'{data_path}/mouse/GSE122187_CellStateKernels.xls.gz') as fin:
    kernels = pd.read_excel(fin, skiprows=1).set_index('Gene')

kernels = kernels.loc[
    :, kernels.columns.sort_values()
]

kernel_name_map = {
    0: "trophoblast stem cells",
    1: "neural ectoderm anterior",
    2: "primitive streak late",
    3: "anterior primitive streak",
    4: "primitive/definitive endoderm",
    5: "allantois",
    6: "secondary heart field/splanchnic lateral plate",
    7: "gut endoderm",
    8: "ectoderm early 1",
    9: "primitive blood early",
    10: "preplacodal ectoderm",
    11: "neural ectoderm posterior",
    12: "posterior lateral plate mesoderm",
    13: "hematopoietic/endothelial progenitors",
    14: "parietal endoderm",
    15: "amnion mesoderm early",
    16: "surface ectoderm",
    17: "epiblast",
    18: "somites",
    19: "ectoderm early 2",
    20: "splanchnic lateral plate/anterior paraxial mesoderm",
    21: "primitive heart tube",
    22: "primitive blood late",
    23: "notochord",
    24: "fore/midbrain",
    25: "distal extraembryonic ectoderm",
    26: "neuromesodermal progenitor early",
    27: "primordial germ cells",
    28: "differentiated trophoblasts",
    29: "visceral endoderm early",
    30: "presomitic mesoderm",
    31: "neuromesodermal progenitor late",
    32: "angioblasts",
    33: "neural crest",
    34: "pharyngeal arch mesoderm",
    35: "similar to neural crest",
    36: "primitive blood progenitors",
    37: "primitive streak early",
    38: "node",
    39: "future spinal cord",
    40: "visceral endoderm late",
    41: "amnion mesoderm late",
}

kernel_name_map = {k: v.capitalize() for k, v in kernel_name_map.items()}
kernels.columns = [kernel_name_map[x] for x in kernels.columns]

print('read kernels')

gene_expression_df = pd.read_csv(f'{data_path}/preprocessed/GSE117542_embryo2_normalized_log_counts.txt')
counts = gene_expression_df.T
counts_z = counts \
    .subtract(counts.mean(axis=1), axis=0) \
    .divide(counts.std(axis=1), axis=0)
print('standardized counts')

pca_data = PCA(n_components=20).fit_transform(counts_z.values.T)
print('PCA done')
umap_data = UMAP(n_neighbors=30, min_dist=.5).fit_transform(pca_data)
print('umap done')
umap_data = pd.DataFrame(umap_data, index=counts.columns, columns=['UMAP1', 'UMAP2'])

common = counts.index & kernels.index
# HotSpot tutorial used consine dist, but dataset paper supplement says to use
# euclidean dist, so we use euclidean too.
dd = cdist(counts.loc[common].T, kernels.loc[common].T, metric='euclidean')
print('cdist done')
dd = pd.DataFrame(dd, index=counts.columns, columns=kernels.columns)
assignment = dd.idxmin(axis=1).loc[counts.columns]
print('assignment done')

colors = sns.color_palette("husl", assignment.unique().size)
plt.figure(figsize=(9, 9))
for i, ct in enumerate(assignment.unique()):
    ii = assignment.index[assignment == ct]
    plt.plot(umap_data.loc[ii].UMAP1, umap_data.loc[ii].UMAP2,
            'o', ms=2, label=ct, color=colors[i])

plt.xticks([])
plt.yticks([])
plt.xlabel('UMAP-1')
plt.ylabel('UMAP-2')
plt.subplots_adjust(right=.6, bottom=.5)
plt.legend(bbox_to_anchor=[1, 1], loc='upper left', markerscale=3);
```

```python
cell_types = pd.DataFrame(assignment)
cell_types.to_csv(f'{data_path}/preprocessed/GSE117542_embryo2_cell_types.txt', header=False, index=False)
cell_types[0].value_counts()
```
